**Paso 1: Generar Datos de Ejemplo**

Primero, generamos un conjunto de datos sint√©ticos.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sint√©ticos
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Datos sint√©ticos')
plt.show()
```

**Paso 2: Implementar el Algoritmo de Gradiente Descendente**

Implementamos el algoritmo de Gradiente Descendente para minimizar la funci√≥n de costo de la regresi√≥n lineal.

```python
# Inicializaci√≥n de par√°metros
m = 0
b = 0
learning_rate = 0.01
n_iterations = 1000
n = len(X)

# Funci√≥n de costo
def compute_cost(X, y, m, b):
    y_pred = m * X + b
    cost = (1 / (2 * n)) * np.sum((y_pred - y) ** 2)
    return cost

# Gradiente Descendente
cost_history = []

for i in range(n_iterations):
    y_pred = m * X + b
    dm = (1 / n) * np.sum(X * (y_pred - y))
    db = (1 / n) * np.sum(y_pred - y)

    m -= learning_rate * dm
    b -= learning_rate * db

    cost = compute_cost(X, y, m, b)
    cost_history.append(cost)

    if i % 100 == 0:
        print(f'Iteraci√≥n {i+1}: m = {m}, b = {b}, Costo = {cost}')

# Par√°metros finales
print(f'Par√°metros finales: m = {m}, b = {b}')
```

**Paso 3: Visualizaci√≥n de la Convergencia**

Podemos visualizar c√≥mo la funci√≥n de costo disminuye con el tiempo.

```python
plt.plot(range(n_iterations), cost_history)
plt.xlabel('Iteraciones')
plt.ylabel('Costo')
plt.title('Convergencia de Gradiente Descendente')
plt.show()
```

**Paso 4: Visualizaci√≥n del Ajuste Final**

Finalmente, podemos visualizar c√≥mo la l√≠nea ajustada se compara con los datos originales.

```python
plt.scatter(X, y, label='Datos')
plt.plot(X, m * X + b, color='red', label='Ajuste lineal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Ajuste lineal usando Gradiente Descendente')
plt.show()
```
