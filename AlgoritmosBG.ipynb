```python
#Algoritmos
#Algoritmo de Búsqueda por fuerza bruta

#Este enfoque consiste en probar todas las posibles soluciones hasta encontrar la correcta. Un ejemplo clásico de búsqueda por fuerza bruta es el problema de encontrar un elemento en una lista.

#Supongamos que queremos encontrar si un número específico está presente en una lista de números. Aquí está el código para hacer esto usando fuerza bruta:

def brute_force_search(arr, target):
    for index, element in enumerate(arr):
        if element == target:
            return index  # Retorna el índice donde se encuentra el elemento
    return -1  # Retorna -1 si el elemento no se encuentra en la lista

# Lista de números
numbers = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

# Número a buscar
target = 9

# Ejecución del algoritmo de búsqueda por fuerza bruta
result = brute_force_search(numbers, target)

if result != -1:
    print(f"El número {target} se encuentra en el índice {result}.")
else:
    print(f"El número {target} no se encuentra en la lista.")


```

```python
#Genetico - F(x) = x^2

#Explicación
#fitness_function(x): Evalúa qué tan buena es una solución (en este caso, simplemente  F(x) = x^2).
#create_individual(): Genera un individuo aleatorio (un número entre -10 y 10).
#create_population(size): Crea una población inicial de tamaño dado.
#select_individual(population): Selecciona un individuo mediante torneo, eligiendo al mejor entre un subconjunto aleatorio de la población.
#crossover(parent1, parent2): Combina dos individuos para crear un nuevo individuo.
#mutate(individual): Aplica mutación a un individuo con cierta probabilidad.
#genetic_algorithm(population_size, generations): Implementa el algoritmo genético principal.


import random

# Definición de la función objetivo
def fitness_function(x):
    return x**2

# Crear un individuo aleatorio
def create_individual():
    return random.uniform(-10, 10)

# Crear una población inicial
def create_population(size):
    return [create_individual() for _ in range(size)]

# Selección por torneo
def select_individual(population):
    tournament_size = 3
    tournament = random.sample(population, tournament_size)
    tournament.sort(key=fitness_function, reverse=True)
    return tournament[0]

# Cruce de dos individuos
def crossover(parent1, parent2):
    return (parent1 + parent2) / 2

# Mutación de un individuo
def mutate(individual):
    mutation_rate = 0.1
    if random.random() < mutation_rate:
        return individual + random.uniform(-1, 1)
    return individual

# Algoritmo genético
def genetic_algorithm(population_size, generations):
    population = create_population(population_size)

    for generation in range(generations):
        new_population = []

        for _ in range(population_size):
            parent1 = select_individual(population)
            parent2 = select_individual(population)
            offspring = crossover(parent1, parent2)
            offspring = mutate(offspring)
            new_population.append(offspring)

        population = new_population
        best_individual = max(population, key=fitness_function)
        print(f"Generación {generation+1}: Mejor solución = {best_individual}, Fitness = {fitness_function(best_individual)}")

    return best_individual

# Parámetros del algoritmo
population_size = 20
generations = 50

# Ejecutar el algoritmo genético
best_solution = genetic_algorithm(population_size, generations)
print(f"Mejor solución encontrada: x = {best_solution}, f(x) = {fitness_function(best_solution)}")
```

```python
#Vamos a implementar el recocido simulado para encontrar el valor máximo de la función (f(x) = -x^2 + 4x)
#Explicación
#objective_function(x): Evalúa qué tan buena es una solución (en este caso, (f(x) = -x^2 + 4x)).
#neighbor(x): Genera una solución vecina agregando un pequeño valor aleatorio a la solución actual.
#cooling_schedule(t, alpha): Define cómo se reduce la temperatura. En este caso, se multiplica por un factor α.
#simulated_annealing(start_temp, alpha, min_temp, max_iterations): Implementa el algoritmo de recocido simulado.
#   start_temp: Temperatura inicial.
#   alpha: Factor de enfriamiento.
#   min_temp: Temperatura mínima.
#   max_iterations: Número máximo de iteraciones.


import math
import random

# Definición de la función objetivo
def objective_function(x):
    return -x**2 + 4*x

# Generar un vecino de la solución actual
def neighbor(x):
    return x + random.uniform(-1, 1)

# Programa de enfriamiento
def cooling_schedule(t, alpha):
    return t * alpha

# Función de recocido simulado
def simulated_annealing(start_temp, alpha, min_temp, max_iterations):
    current_solution = random.uniform(-10, 10)
    current_value = objective_function(current_solution)
    temperature = start_temp

    while temperature > min_temp and max_iterations > 0:
        # Generar un vecino
        new_solution = neighbor(current_solution)
        new_value = objective_function(new_solution)

        # Decidir si se acepta el vecino
        if new_value > current_value or random.random() < math.exp((new_value - current_value) / temperature):
            current_solution = new_solution
            current_value = new_value

        # Reducir la temperatura
        temperature = cooling_schedule(temperature, alpha)
        max_iterations -= 1

        print(f"Temperatura: {temperature:.2f}, Solución: {current_solution:.4f}, Valor: {current_value:.4f}")

    return current_solution, current_value

# Parámetros del algoritmo
start_temp = 100  # Temperatura inicial
alpha = 0.9  # Factor de enfriamiento
min_temp = 0.01  # Temperatura mínima
max_iterations = 1000  # Máximo número de iteraciones

# Ejecutar el algoritmo de recocido simulado
best_solution, best_value = simulated_annealing(start_temp, alpha, min_temp, max_iterations)

print(f"Mejor solución encontrada: x = {best_solution}, f(x) = {best_value}")


#El algoritmo comienza con una temperatura alta que permite aceptar soluciones peores para evitar mínimos locales.
#A medida que la temperatura disminuye, el algoritmo se vuelve más selectivo, prefiriendo soluciones que mejoran la función objetivo.
#El proceso continúa hasta que la temperatura cae por debajo de un umbral mínimo o se alcanza el número máximo de iteraciones.

#Este enfoque es útil para problemas de optimización complejos donde los métodos de búsqueda tradicionales podrían quedarse atrapados
#en óptimos locales.
```
