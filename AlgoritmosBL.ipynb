```python
#Algoritmos
#Algoritmo de Búsqueda en la profundidad

# Representación de un grafo usando una lista de adyacencia
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Función de búsqueda en profundidad (DFS) utilizando recursión
def dfs_recursive(graph, start, visited=None):
    if visited is None:
        visited = set()

    visited.add(start)
    print(start, end=' ')  # Procesar el nodo (en este caso, imprimirlo)

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs_recursive(graph, neighbor, visited)

# Llamada a la función DFS comenzando desde el nodo 'A'
print("Recorrido DFS (recursivo):")
dfs_recursive(graph, 'A')
print()  # Para nueva línea

# Función de búsqueda en profundidad (DFS) utilizando una pila
def dfs_stack(graph, start):
    visited = set()
    stack = [start]

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            print(node, end=' ')  # Procesar el nodo (en este caso, imprimirlo)
            # Agregar los vecinos a la pila en orden inverso para mantener el orden de visita
            for neighbor in reversed(graph[node]):
                if neighbor not in visited:
                    stack.append(neighbor)

# Llamada a la función DFS utilizando una pila comenzando desde el nodo 'A'
print("Recorrido DFS (usando pila):")
dfs_stack(graph, 'A')
print()  # Para nueva línea
```

```python
#Búsqueda en Anchura

from collections import deque

# Representación de un grafo usando una lista de adyacencia
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

# Función de búsqueda en anchura (BFS)
def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            print(node, end=' ')  # Procesar el nodo (en este caso, imprimirlo)
            for neighbor in graph[node]:
                if neighbor not in visited:
                    queue.append(neighbor)

# Llamada a la función BFS comenzando desde el nodo 'A'
print("Recorrido BFS:")
bfs(graph, 'A')
print()  # Para nueva línea
```

```python
#Hill Climbing
#f(x) = -x^2 + 4x

import random

# Definición de la función objetivo
def objective_function(x):
    return -x**2 + 4*x

# Función de Hill Climbing
def hill_climbing(start, step_size, max_iterations):
    current_solution = start
    current_value = objective_function(current_solution)

    for _ in range(max_iterations):
        # Generar vecinos
        neighbors = [current_solution - step_size, current_solution + step_size]

        # Evaluar vecinos
        next_solution = max(neighbors, key=objective_function)
        next_value = objective_function(next_solution)

        # Si no hay mejora, detener la búsqueda
        if next_value <= current_value:
            break

        # Moverse al mejor vecino
        current_solution = next_solution
        current_value = next_value

        print(f"Iteración: {_}, Solución: {current_solution}, Valor: {current_value}")

    return current_solution, current_value

# Parámetros del algoritmo
start = random.uniform(-10, 10)  # Punto de inicio aleatorio
step_size = 0.1  # Tamaño del paso para generar vecinos
max_iterations = 100  # Máximo número de iteraciones

# Ejecución del algoritmo Hill Climbing
best_solution, best_value = hill_climbing(start, step_size, max_iterations)

print(f"Mejor solución encontrada: x = {best_solution}, f(x) = {best_value}")
```
